# Making plots for Resized Images

```python
sns.set(rc={'figure.figsize':(12,8)})
# sns.countplot(x = df['label'], palette= 'Spectral', order=df['label'].value_counts().index[::-1])
sns.countplot(x = df['label'], palette= 'Spectral', order=df['label'].value_counts().index)
plt.title('Number of Images per Class')
plt.show()
```

```python
sns.set(rc={'figure.figsize':(12,8)})
sns.histplot(data=df, x = 'KB_size', palette='Spectral', bins=15)
plt.xlabel('Image memory in KB')
plt.title('Histogram of Image memory usage (400x400 images)')
plt.show()
```

```python
df2 = df.groupby(['patient_id'])['patient_id'].count()
```

```python
sns.set(rc={'figure.figsize':(12,8)})
sns.countplot(x = df2, palette= 'Spectral')
plt.title('Number of Images per Patient')
plt.show()
```

```python
sns.set(rc={'figure.figsize':(12,8)})
sns.boxplot(data=df, x = 'center_id', y = 'new_KB_size', palette = 'Spectral')
plt.title('Boxplots of Image sizes per Center_id')
plt.ylabel('KB size')
plt.show()
```

```python
sns.set(rc={'figure.figsize':(12,8)})
sns.countplot(data= df, x='center_id', palette= 'Spectral')
plt.title('Number of Images per center_id')
plt.show()
```

```python
sns.set(rc={'figure.figsize':(12,8)})
sns.countplot(data= df, x='center_id', hue='label', palette= 'Spectral')
plt.title('Number of Images per center_id and Class')
plt.show()
```
